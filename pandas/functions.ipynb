# Funciones principales para DS de Pandas

```python
import pandas as pd
```

```python
df_books = pd.read_csv('./data_sources/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv', sep=',', header=0)
```

## head()
Muestra los primeros rows del DataFrame, por default retorna los primeros 5

```python
df_books.head()
```

```python
df_books.head(1)
```

## tail()
Es la contraparte de head, retorna los ultimos rows del DataFrame, por default retorna los ultimos 5 rows.

```python
df_books.tail()
```

```python
df_books.tail(1)
```

## info()
Regresa la información del DataFrame

```python
df_books.info()
```

Nos regresa El número total de filas y columnas, desde y hasta donde esta el index.
<br> 

En las columnas nos indica si hay valores nullos el tipo de dato que contiene cada columna, y el resumen de cuantas columnas pertenecen a cada tipo de dato y al final el total de memoria que utiliza el DataFrame.

## describe()
Muestra un DataFrame nuevo con valores estadisticos de aquellas columnas numéricas.

```python
df_books.describe()
```

Con esta función podemos ver desde el principio de un analisis cual es la distribución de nuestros datos.

## memory_usage()
Nos indica la memoria en disco que esta utilizando el DataFrame.

Por default no recibe ningún parámetro, pero si le enviamos `deep` con el valor en `True` retorna un análisis de memoria mas certero, donde dice especificamente cuanta memoria ocupa cada columna del DataFrame.

```python
df_books.memory_usage(deep=True)
```

## value_counts()
Retorna el total de veces que aparecen los valores por columna en todo en un DataFrame.

Obtentremos las veces que aparecen los autores en `df_books`

```python
df_books['Author'].value_counts()
```

## drop_duplicates()
Elimina los valores duplicados a nivel de row.

Dumplicamos un row

```python
df_books = pd.concat((df_books, df_books.iloc[0].to_frame().T))
df_books
```

```python
df_books.drop_duplicates()
```

Si de los duplicados queremos conservar el ultimo, podemos enviar el parámetro `keep` con el valor `last`

```python
df_books.drop_duplicates(keep='last')
```

## sort_values()
Ordena el DataFrame por columna, por default el ordenamiento lo hace de forma acendente.

```python
df_books.sort_values('Year')
```

Para ordernar de forma descentente se lo indicamos en el parámetro `ascending` con el valor en `False`

```python
df_books.sort_values('Year', ascending=False)
```
