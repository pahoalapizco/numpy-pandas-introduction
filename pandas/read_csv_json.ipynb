# Leer archivos CSV y JSON con Pandas

La libreria Pandas esta enfocada en la manipulación y analísis de datos, una de sus particularidades es que nos permite leer archivos externos de tipo CSV, JSON, HTML, XML, y un largo etc. Y trabajar posteriormente con su contenido.

## CSV

Por sus siglas en ingles "**C**oma **S**epareted **V**alues", estos son archivos de formato abierto que representan los datos en formato de tablas, donde cada valor esta separado por comas (,) y cada separación corresponde a una columna.
<br>
Este es el formato más común para trabajar con datos en el mundo de Data Science.

```python
files_path = '/home/pahoalapizco/anaconda3/curso-pandas-numpy/pandas/data_sources'
```

```python
csv_path = f'{files_path}/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv'
csv_path
```

Al leer archivos CSV con pandas podemos trabajar con él en una estructura matricial.

```python
df_books = pd.read_csv(csv_path)
df_books
```

La función `read_csv` recibe como parámetro obligatorio la ruta del archivo que va a leer, pero también recibe otro parametros que son opcionales, algunos de ellos son los siguientes:

#### sep
En este parametro indicamos como estan separados los valores del archivo, pandas toma por default la coma (,) siendo este el más común, pero hay ocasiones que podemos encontrarnos un pipe (|) como delimitador de columnas, es en ese caso cuando debemos especificarlo.

```python
df_books = pd.read_csv(csv_path, sep='|') # Al indicale un separador diferente al que lleva el archivo, la estructura que obtenemos cambia (no es la correcta).
df_books
```

#### header
Le indicamos que fila del archivo CSV contiene el encabezado. Por default, toma el 0.

```python
df_books = pd.read_csv(csv_path, sep=',', header=2)
df_books
```

### names
Lista de los nombres de las columnas del archivo (array de strings), si el archivo no lleva nombre de encabezados definidos o bien si queremos sobre escribir los ya existentes

```python
df_books = pd.read_csv(csv_path, names=['Nombre', 'Autor', 'User Rating', 'Reviews', 'Price', 'Year', 'Genre'])
df_books
```

[Documentación de read_csv](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html)

## JSON
Por sus siglas en ingles **J**ava**S**cript **O**bject **N**otation, es un archivo de datos semiestructurado, facil de leer por su formato de clave-valor.

```python
json_path = f'{files_path}/HPCharactersDataRaw.json'
```

```python
df_characters = pd.read_json(json_path)
df_characters
```

Al igual que con los archivos CSV con los archivos JSON pandas nos permite trabajar con ellos en una estructura matricial, tomando como nombres de columnas las claves/llaves del documento.
<br>
También podemos verlo y trabajar con el en su formato raw, sin embargo esa forma de trabajarlo no es la más común en Data Science.

```python
df_characters = pd.read_json(json_path, typ='Series')
df_characters
```

[Documentación de read_json](https://pandas.pydata.org/docs/reference/api/pandas.read_json.html)
