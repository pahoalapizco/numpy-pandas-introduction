# Groupby

Similar al agrupamiento de SQL pero con Python y Pandas. Esta función nos permite agrupar los datos de un DataFrame y aplicarle funciones de agregación.

```python
import pandas as pd
```

```python
df_books = pd.read_csv('./data_sources/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv', sep=',', header=0)
df_books.head()
```

A la función `groupby` podemos enviar por parametro el `label` (nombre de la columna) por la cual queremos agrupar o bien una lista con las columnas a agrupar.
<br>
Por si sola esta función no nos muestra los datos, para ello le aplicamos funciones de agregación según lo que queremos explorar.

```python
groupby_author = df_books.groupby('Author')
groupby_author
```

### count()
Total de rows agrupados por columna.

```python
groupby_author.count()
```

### median()
Calcula la media de las columnas numéericas.

```python
groupby_author.median()
```

### sum()
Calcula la sumatoria de los valores numéericos que fueron agrupados.

```python
groupby_author.sum()
```

### agg()
Permite agregar varias funciones de agregación a la vez a un mismo conjunto de datos agrupados.

#### min y max

```python
groupby_author.agg(['min', 'max'])
```

Con `agg` podemos ser más especificos e indicarle exactamente a que columna aplicarle cierta función de agregación.

Obtener el rating maximo y minimo, sumar el user rating.

```python
groupby_author.agg({'Reviews': ['min', 'max'], 'User Rating': 'sum'})
```

## Agrupar por más de una columna

```python
groupby_author_year = df_books.groupby(['Author', 'Year'])
```

```python
groupby_author_year.count()
```

```python
groupby_author_year.sum()
```

Al aplicar `groupby` a un DataFrame, las columnas por las cuales agrupamos se convierten en indices de un nuevo DF, por esa razon los títulos 'Author' y 'Year' en el ejemplo anterior no aparecen al mismo nivel que las columnas "ser Rating", "Reviews" y "Price".
<br>
Por lo tanto al ser indices podemos acceder a sus datos con la función `loc`

```python
groupby_author_year.sum().loc['Stephen King']
```

```python
groupby_author_year.sum().loc['Stephen King': 'year']
```

Con `reset_index()` podemos subir a nivel de columnas los indices que el agrupado generó .

```python
groupby_author_year.sum().reset_index()
```
